# Seabornによるデータの可視化

### ライブラリのインポート

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

import japanize_matplotlib
import warnings
warnings.filterwarnings('ignore')
```

### データの読み込み

```python
iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
df_iris = pd.DataFrame(iris)
df_titanic = pd.DataFrame(titanic)
```

```python
df_iris
```

## 基本グラフ操作

### scatterplot (散布図)

```python
sns.scatterplot(x="sepal_length", y="sepal_width", hue="species", data=df_iris)
plt.show()
```

### relplot（scatterplotを複数配置できるグラフ)
* カテゴリ区分の引数:row='', col=''も指定できる

```python
sns.relplot(x="sepal_length", y="sepal_width", hue="species", data=df_iris)
plt.show()
```

```python

sns.relplot(x="sepal_length", y="sepal_width", hue="species", col="species", data=df_iris)
```

### stripplot（1変数の散布図)

```python
sns.stripplot(x="species", y="sepal_length", hue="species", data=df_iris)
```

### catplot（stripplotを複数配置できるグラフ)
* カテゴリ区分の引数:row='', col=''を指定

```python
sns.catplot(x="species", y="sepal_length", hue="species", col="species", data=df_iris)
```

### swarmplot（点を重ねない1変数の散布図)

```python
sns.swarmplot(x="species", y="sepal_length", hue="species", data=df_iris)
```

### regplot（線形回帰モデルのグラフ)

```python
sns.regplot(x='sepal_length', y='sepal_width', data=df_iris)
```

### lmplot（regplotを複数配置できるグラフ)
* 引数にcol, rowを追加

```python
sns.lmplot(x='sepal_length', y='sepal_width', col='species', hue='species', data=df_iris)
```

### histplot（ヒストグラム）
* distplotは非推奨とのこと(カーネル密度関数も出してくれる)

```python
sns.histplot(df_iris, x='sepal_length', hue='species')
```

```python
sns.distplot(df_iris['sepal_length'], kde=False)
```

### displott（histplotを複数配置できるグラフ）
* 引数にcol, rowを追加

```python
sns.displot(df_iris, x='sepal_length', col='species')
```

### ecdfplot（累積分布関数)

```python
sns.ecdfplot(x='sepal_length', data=df_iris)
```

### jointplot（ヒストグラムつき散布図)

```python
sns.jointplot(x='sepal_length', y='sepal_width', data=df_iris, hue='species')
```

### kdeplot（カーネル密度推定を使用した分布)

```python
sns.kdeplot(x='sepal_length', hue='species', data=df_iris)
```

### barplot（棒グラフ)

```python
sns.barplot(x='species', y='sepal_width', hue='species', data=df_iris)
```

### catplot（barplotを複数配置できるグラフ)
* 引数にカテゴリ変数col, rowを追加

```python
sns.catplot(kind="bar", x='species', y='sepal_width', hue='species', col='species', data=df_iris)
```

### countplot（カテゴリ別のカウントグラフ）
* hue='species'

```python
sns.countplot(x='species', data=df_iris)
```

### boxplot（箱ひげ図)
* カテゴリ引数:x, hue

```python
sns.boxplot(x='species', y='sepal_width', data=df_iris)
```

### violinplot（バイオリン図）
* カテゴリ引数:x, hue

```python
sns.violinplot(x='species', y='sepal_width', split=True, data=df_iris)
```

### pointplot（平均値と信頼区間のグラフ）
* カテゴリ引数:x, hue

```python
sns.pointplot(x='species', y='sepal_width', data=df_iris)
```

### lineplot（折れ線グラフ）

```python
sns.lineplot(data=df_iris)
```

```python
sns.relplot(x='sepal_length', y='sepal_width', data=df_iris, kind='line')
```

### heatmap（ヒートマップ）

```python
sns.heatmap(df_iris.corr(), annot=True, cmap='bwr', fmt='.2f', vmax=1, vmin=-1, center=0)
```

### catplotの種類(引数によって変更できる)
* kindなし
* kind='bar'
* kind='box'

```python
sns.catplot(kind='box', x="species", y="sepal_length", hue="species", data=df_iris)
```

## グラフ種類ごとの設定

#### 折れ線グラフ

* sns.lineplot()

|引数|説明|
|----|----|
|x, y（省略可）|データフレームのカラム名|
|data|データフレーム|
|hue|カラムのカテゴリを異なる色で描画|
|style|カラムのカテゴリを異なる線種で描画|
|ci|信頼区間の幅（デフォルト=95）で半透明のエラーバンドを表示, 表示しない場合はNone|
|ax|グラフを描画したい領域（axesオブジェクトでの指定）|

```python
sns.lineplot(x='sepal_length', y='sepal_width', data = df_iris, ci=None)
```

### 棒グラフ

* sns.barplot()

| 引数 | 説明 |
| ---- | ---- |
|x, y（省略可） | データフレームのカラム名（xに数値型の変数、yにカテゴリ変数を指定した場合は横棒グラフを描画）|
|data|データフレーム |
|order|表示する並び（リストで指定） |
|hue|カラムのカテゴリを異なる色で描画 |
|ci|信頼区間の幅（デフォルト=95）で半透明のエラーバンドを表示, 表示しない場合はNone|
|ax|グラフを描画したい領域（axesオブジェクトでの指定）

* y軸にカテゴリ変数を指定した場合には、横棒グラフになる
* カテゴリの並び順はorder=['a', 'b', 'c']とリストで指定

```python
sns.barplot(x='species', y='sepal_length', data=df_iris)
```

### 散布図

* sns.scatterplot()

|引数 | 説明|
|----|----|
|x, y |データフレームのカラム名|
|data |データフレーム|
|hue |カラムのカテゴリを異なる色で描画|
|style |カラムのカテゴリを異なる線種で描画|
|size |カラムのカテゴリを異なるサイズで描画|
|alpha |透明度|
|ax |グラフを描画したい領域（axesオブジェクトでの指定）|



```python
sns.scatterplot(x='sepal_length', y='sepal_width', data=df_iris, style='species')
```

### 箱ひげ図

* sns.boxplot()

|引数|説明|
|----|----|
|x, y（省略可）| 描画対象のデータのカラム名|
|data| 描画対象のデータフレーム|
|order| 出力する順番（文字列のリストで指定）|
|hue| 指定したカラムのカテゴリでデータを分割し、グラフを描画|
|ax| グラフを描画したい領域（axesオブジェクトでの指定）|




```python
sns.boxplot(x='species', y='sepal_width', data=df_iris)
```

### ヒストグラム

* sns.histplot()

|引数| 説明|
|----|----|
|x, y（省略可）| 描画対象のデータのカラム名|
|data| 描画対象のデータフレーム|
|bin| ビンの分割|
|stat| 各ビンで計算する統計法 <br>・count <br>・frequency <br>・density <br>・probability|
|order| 出力する順番（文字列のリストで指定）|
|hue| 指定したカラムのカテゴリでデータを分割し、グラフを描画|
|ax| グラフを描画したい領域（axesオブジェクトでの指定）|

* 積上げヒストグラム : multiple='stack'
* カーネル密度推定の表示 : kde=True

```python
sns.histplot(x='sepal_width', bins=10, data=df_iris, hue='species')
```

### ヒートマップ

* sns.heatmap()

|引数 | 説明|
|----|----|
|x, y（省略可）| 描画対象のデータのカラム名|
|data| データフレーム|
|bin| ビンの分割|
|annot| 数値の表示|
|linewidths| 各セルを分割する線の幅|
|ax| グラフを描画したい領域（axesオブジェクトでの指定）|

```python
sns.heatmap(df_iris.corr(), annot=True, cmap='bwr', fmt='.2f', vmax=1, vmin=-1, center=0)
```

## グラフの詳細設定

### 軸、目盛りの設定
* 軸範囲は matplotlib.pyplot の xlim (開始, 終了) , ylim (開始, 終了)を設定

```python
sns.scatterplot(x='sepal_length', y='sepal_width', data=df_iris, style='species')
plt.title('title')
plt.xlabel('xlabel')
plt.ylabel('ylabel')
plt.ylim(0,10)

plt.show()
```

### グラフに縦・横線を追加
* plt.vlines([線を引きたい値], 線の小さい側の値, 線の大きい側の値, 色, 線のスタイル)
* plt.hlines([線を引きたい値], 線の小さい側の値, 線の大きい側の値, 色, 線のスタイル)

```python
sns.displot(x='sepal_width', bins=10, data=df_iris, hue='species')
plt.xlim(0,6)
plt.vlines(5, 0, 16, 'red', linestyles='dashed')
```

### 凡例の表示/非表示
* グラフの表示メソッドの引数に legend=False を追加
